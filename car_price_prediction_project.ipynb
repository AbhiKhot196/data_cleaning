# Work on year Column 

```python
ds['year'].unique()
```

```python
ds=ds[ds['year'].str.isnumeric()] # only take int no from column
ds['year'].unique() # check str value remove or not 
```

```python
ds['year']=ds['year'].astype(int) # Change data type of year column 
```

# work on Price column 

```python
ds=ds[ds['Price']!='Ask For Price'] # take value price not equal to 'ask for price'
```

```python
ds['Price']=ds['Price'].str.replace(',','') # Removing Comma from column
```

# work on kms_driven column

```python
ds
```

```python
ds['kms_driven']=ds['kms_driven'].str.split().str.get(0) # remove kms object from column 
```

```python
ds['kms_driven']=ds['kms_driven'].str.replace(',','') # remove or replace comma from column
```

```python
ds
```

#  Work on name column

```python
ds['name'].unique()
```

```python
ds['name']=ds['name'].str.split().str.get(1) # remove compony name from car names 
ds
```

```python
ds.dropna()
```

```python
ds.drop_duplicates(inplace=True)
```
